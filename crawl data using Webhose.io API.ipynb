```python
#Implementation:get data from Webhose.io queries

import webhoseio

webhoseio.config(token="6c6eb1dd-e403-403b-a853-28b501c61877")
query_params = {
    "q": "language:english organization:IBM",
    "ts": "1566166830669",
    "sort": "published"
    }
output = webhoseio.query("filterWebContent", query_params)

print(len(output['posts']))
type(output['posts'])
```

```python
#get data from Webhose.io queries 100 times and store query output in feeds

feeds = []
count = 99
while count > 0:    
    output = webhoseio.get_next()
    for item in output['posts']:
        feeds.append(item)
        #print(item['title'], item['published'])
    count -= 1
print(len(feeds))
```

```python
print(len(feeds))
```

```python
#Stroage: write query data from feeds to JSON file
import json
with open("IBM_english.json", "w") as myfile:
    for feed in feeds:
        line = json.dumps(feed)
        myfile.write(line)
        myfile.write("\n")
        
```

```python
#read data from json file
json_data=open("IBM_english.json").readlines()
#print(json_data)

#Output: print the number of articles, titles and publish dates of the first 100 articles.
feeds_read_from_file = []
for line in json_data:
    feeds_read_from_file.append(json.loads(line))
print(len(feeds_read_from_file))

for feed in feeds_read_from_file[0:99]:
    print(feed['title'],feed['published'])
```
