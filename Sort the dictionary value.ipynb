```python
#method 1 using Using Python sorted function on the dictionary sorted() 
import csv
input_file=open('/Users/rissacao/Downloads/cb_sample.csv','r')
reader=csv.reader(input_file)

data={}
#skip header(first row)
headers = next(reader)
for row in reader:
    data[row[0]]={'city':row[1],'region':row[2],'country':row[3],'category_list':row[4],'category_list':row[5],'cb_url':row[6],'rank':row[7],'domain':row[8]}

print(len(data))

# using Using Python sorted function on the dictionary sorted() 
# Sort nested dictionary by key 
res = sorted(data.items(), key = lambda x: x[1]['rank']) 
print(res)
print(len(res))
```

```python
#method 2 Copying values into an array or list, and sorting
import csv
input_file = csv.DictReader(open("/Users/rissacao/Downloads/5430/cb_sample.csv"))

#Copying values into an array or list, and sorting
list_to_be_sorted=list(input_file)

print(type(list_to_be_sorted))
print(list_to_be_sorted)
```

```python
#Sort the dictionary by the "rank" field in ascending order:
newlist = sorted(list_to_be_sorted, key=lambda k: k['rank'])
print(len(newlist))
print(newlist)
```

```python
#convert a list of nested OrderedDict to a list of dict
from json import loads, dumps
from collections import OrderedDict

def to_dict(input_ordered_dict):
    return loads(dumps(input_ordered_dict))

result=to_dict(newlist)
print(result)
len(result)
```

```python
#Create a set containing only unique (non-repeated) rows from the list of dictionaries
unique_list = []
seen_items = set()

for dictionary in result:
    t = tuple(dictionary.items())
    if t not in seen_items:
        seen_items.add(t)
        # duplicate-check passed, adding the unique key back to it's dictionry
        unique_list.append(dictionary)

print(unique_list) 
print(len(unique_list))
```
